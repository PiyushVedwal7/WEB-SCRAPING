```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import certifi

# Function to extract Product Title
def get_title(soup):
    try:
        title = soup.find("span", attrs={"id": 'productTitle'})
        title_string = title.text.strip()
    except AttributeError:
        title_string = ""
    return title_string

# Function to extract Product Price
def get_price(soup):
    try:
        price = soup.find("span", attrs={'id': 'priceblock_ourprice'}).string.strip()
    except AttributeError:
        try:
            price = soup.find("span", attrs={'id': 'priceblock_dealprice'}).string.strip()
        except:
            price = ""
    return price

# Function to extract Product Rating
def get_rating(soup):
    try:
        rating = soup.find("i", attrs={'class': 'a-icon a-icon-star a-star-4-5'}).string.strip()
    except AttributeError:
        try:
            rating = soup.find("span", attrs={'class': 'a-icon-alt'}).string.strip()
        except:
            rating = ""
    return rating

# Function to extract Number of User Reviews
def get_review_count(soup):
    try:
        review_count = soup.find("span", attrs={'id': 'acrCustomerReviewText'}).string.strip()
    except AttributeError:
        review_count = ""
    return review_count

# Function to extract Availability Status
def get_availability(soup):
    try:
        available = soup.find("div", attrs={'id': 'availability'}).find("span").string.strip()
    except AttributeError:
        available = "Not Available"
    return available

# Function to extract Product About Section
def about(soup):
    try:
        about = soup.find("span", attrs={'a-list-item'}).find("span").string.strip()
    except AttributeError:
        about = "none"
    return about

if __name__ == '__main__':
    # Add your user agent
    HEADERS = ({'User-Agent': '', 'Accept-Language': 'en-US, en;q=0.5'})

    # The webpage URL
    URL = "https://www.amazon.com/s?k=playstation+4&ref=nb_sb_noss_2"

    # HTTP Request (bypassing certificate verification)
    webpage = requests.get(URL, headers=HEADERS, verify=False)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    # Fetch links as List of Tag Objects
    links = soup.find_all("a", attrs={'class': 'a-link-normal s-no-outline'})

    # Store the links
    links_list = [link.get('href') for link in links]

    # Data dictionary
    d = {"title": [], "price": [], "rating": [], "reviews": [], "availability": [], "about": []}

    # Loop for extracting product details from each link
    for link in links_list:
        new_webpage = requests.get("https://www.amazon.com" + link, headers=HEADERS, verify=False)
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        # Function calls to display all necessary product information
        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['reviews'].append(get_review_count(new_soup))
        d['availability'].append(get_availability(new_soup))
        d['about'].append(about(new_soup))

    # Create a DataFrame
    amazon_df = pd.DataFrame.from_dict(d)
    amazon_df['title'].replace('', np.nan, inplace=True)
    amazon_df = amazon_df.dropna(subset=['title'])
    amazon_df.to_csv("amazon_data.csv", header=True, index=False)

    print("Data saved to amazon_data.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import certifi

# Function to extract Job Title
def get_job_title(soup):
    try:
        job_title = soup.find("h3", attrs={"class": "base-search-card__title"}).text.strip()
    except AttributeError:
        job_title = ""
    return job_title

# Function to extract Company Name
def get_company_name(soup):
    try:
        company_name = soup.find("h4", attrs={"class": "base-search-card__subtitle"}).text.strip()
    except AttributeError:
        company_name = ""
    return company_name

# Function to extract Job Link
def get_job_link(soup):
    try:
        job_link = soup.find("a", attrs={"class": "base-card__full-link"})['href']
    except AttributeError:
        job_link = ""
    return "https://www.linkedin.com" + job_link if job_link else ""

# Function to extract Location
def get_location(soup):
    try:
        location = soup.find("span", attrs={"class": "job-search-card__location"}).text.strip()
    except AttributeError:
        location = ""
    return location

if __name__ == '__main__':
    # Headers for LinkedIn request
    HEADERS = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36',
        'Accept-Language': 'en-US, en;q=0.5'
    }

    # LinkedIn Job Search URL (search for Python Developer jobs)
    URL = "https://www.linkedin.com/jobs/search/?keywords=python%20developer"

    try:
        # Make HTTP request with proper certificate verification
        webpage = requests.get(URL, headers=HEADERS, verify=certifi.where())
        
        # Parse the page content using BeautifulSoup
        soup = BeautifulSoup(webpage.content, "html.parser")

        # Find all job cards on the page
        job_cards = soup.find_all("div", attrs={"class": "base-search-card"})

        # Prepare a dictionary to store job details
        job_data = {
            "Job Title": [],
            "Company Name": [],
            "Job Link": [],
            "Location": []
        }

        # Loop through all job cards to extract job details
        for job in job_cards:
            job_data["Job Title"].append(get_job_title(job))
            job_data["Company Name"].append(get_company_name(job))
            job_data["Job Link"].append(get_job_link(job))
            job_data["Location"].append(get_location(job))

        # Print job data
        for i in range(len(job_data["Job Title"])):
            print(f"Job Title: {job_data['Job Title'][i]}")
            print(f"Company Name: {job_data['Company Name'][i]}")
            print(f"Job Link: {job_data['Job Link'][i]}")
            print(f"Location: {job_data['Location'][i]}")
            print("-" * 40)

        # Optionally, save job data to a CSV file
        import pandas as pd
        df = pd.DataFrame(job_data)
        df.to_csv("linkedin_jobs.csv", index=False, header=True)
        print("Job data saved to linkedin_jobs.csv")

    except requests.exceptions.SSLError as e:
        print("SSL Error:", e)
    except OSError as e:
        print("OS Error:", e)
    except Exception as e:
        print("Error:", e)
```

```python
df = pd.DataFrame(job_data)
df.to_excel("linkedin_jobs.xlsx", index=False, header=True, engine='openpyxl')
print("Job data saved to linkedin_jobs.xlsx")
```
